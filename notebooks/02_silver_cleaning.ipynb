```python
bronze_df = spark.table("workspace.security_pipeline.bronze_ids")
silver_df = bronze_df.dropDuplicates()
print("After removing duplicates:", silver_df.count())
```

```python
from pyspark.sql.functions import col, sum

missing_df = silver_df.select([
    sum(col(c).isNull().cast("int")).alias(c) for c in silver_df.columns
])

display(missing_df)
```

```python
from pyspark.sql.types import DoubleType, IntegerType

numeric_cols = [f.name for f in silver_df.schema.fields 
                if isinstance(f.dataType, (DoubleType, IntegerType))]

silver_df = silver_df.fillna(0, subset=numeric_cols)
```

```python
from pyspark.sql.functions import to_timestamp

silver_df = silver_df.withColumn(
    "timestamp",
    to_timestamp("timestamp", "dd/MM/yyyy HH:mm:ss")
)
```

```python
from pyspark.sql.functions import when

silver_df = silver_df.withColumn(
    "label_encoded",
    when(col("label") == "Benign", 0).otherwise(1)
)
```

```python
silver_df.groupBy("label", "label_encoded").count().show()
```

```python
silver_df = silver_df.drop("source_file")  
```

```python
silver_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.security_pipeline.silver_ids")
```
