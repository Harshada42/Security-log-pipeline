```python
from pyspark.sql.functions import current_timestamp, col

raw_path = "/Volumes/workspace/security_pipeline/ids_raw/*.csv"

bronze_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(raw_path)
    .withColumn("source_file", col("_metadata.file_path"))
    .withColumn("ingestion_time", current_timestamp())
)
```

```python
print("Total records:", bronze_df.count())
print("Total columns:", len(bronze_df.columns))

display(bronze_df.limit(5))
```

```python
import re

clean_df = bronze_df

for col_name in clean_df.columns:
    new_col = col_name.strip().lower()
    new_col = re.sub(r"[ /.-]+", "_", new_col)   # replace space, /, . with _
    new_col = re.sub(r"[^a-zA-Z0-9_]", "", new_col)  # remove special chars
    clean_df = clean_df.withColumnRenamed(col_name, new_col)

print(clean_df.columns)
```

```python
clean_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.security_pipeline.bronze_ids")
```
