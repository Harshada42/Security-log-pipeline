```python
%sql
CREATE OR REPLACE TABLE workspace.security_pipeline.gold_label_distribution AS
SELECT 
    label,
    COUNT(*) AS total_records,
    ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM workspace.security_pipeline.silver_ids), 2) AS percentage
FROM workspace.security_pipeline.silver_ids
GROUP BY label;
```

```python
%sql
SELECT * 
FROM workspace.security_pipeline.gold_label_distribution;
```

```python
%sql
CREATE OR REPLACE TABLE workspace.security_pipeline.gold_protocol_distribution AS
SELECT 
    protocol,
    COUNT(*) AS total_records,
    ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM workspace.security_pipeline.silver_ids), 2) AS percentage
FROM workspace.security_pipeline.silver_ids
GROUP BY protocol
ORDER BY total_records DESC;
```

```python
%sql
SELECT * FROM workspace.security_pipeline.gold_protocol_distribution;
```

```python
%sql
CREATE OR REPLACE TABLE workspace.security_pipeline.gold_attack_protocol AS
SELECT 
    label,
    protocol,
    COUNT(*) AS total_records
FROM workspace.security_pipeline.silver_ids
GROUP BY label, protocol
ORDER BY total_records DESC;
```

```python
%sql
SELECT * FROM workspace.security_pipeline.gold_attack_protocol;
```

```python
%sql
CREATE OR REPLACE TABLE workspace.security_pipeline.gold_attack_ports AS
SELECT 
    label,
    dst_port,
    COUNT(*) AS total_records
FROM workspace.security_pipeline.silver_ids
WHERE label != 'Benign'
GROUP BY label, dst_port
ORDER BY total_records DESC;
```

```python
%sql
SELECT *
FROM workspace.security_pipeline.gold_attack_ports
LIMIT 20;
```
