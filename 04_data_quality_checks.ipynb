```python
silver_df = spark.table("workspace.security_pipeline.silver_ids")

print("Silver rows:", silver_df.count())
print("Silver columns:", len(silver_df.columns))
```

```python
total_rows = silver_df.count()
unique_rows = silver_df.dropDuplicates().count()

print("Total rows:", total_rows)
print("Unique rows:", unique_rows)
print("Duplicate rows:", total_rows - unique_rows)
```

```python
silver_df.groupBy("label").count().show()
```

```python
from pyspark.sql.functions import col, sum

important_cols = ["label", "protocol", "dst_port", "timestamp"]

null_check = silver_df.select([
    sum(col(c).isNull().cast("int")).alias(c) for c in important_cols
])

display(null_check)
```
